```python
#Play Rock-Paper-Scissors game
#Takes random value for user
import random

#Some basic info for user
print("Let's start Rock-Paper-Scissors Game")
print("Rules of the game: \n1.Rock vs Scissors = Rock wins\n2.Scissors vs paper = scissors wins\n3.Paper vs rock = paper wins")
print("\n")

#Initialise the variable
user=0
computer=0

#Build the logic
while(True):
    print("\n\nChoice\n1.Rock \n2.Scissors\n3.Paper")
    print("\n")
    ch= int(input("Enter your choice: "))

    while ch>3 or ch<1:
        ch=int(input("Enter a valid choice: "))
    if ch == 1:
        choice = "rock"
    elif ch == 2:
        choice = 'scissors'
    else:
        choice = 'paper'
    print("User choices ",choice)

    print("Now its computer's tern...")
    comp= random.randint(1,3)

    while comp == ch:
        comp=random.randint(1,3)
    if comp==1:
        comp_choice='rock'
    elif comp==2:
        comp_choice='scissor'
    else:
        comp_choice='paper'
    print("Computer choice is ", comp_choice)

    if ch == comp:
        print('Its a Draw',end="")
        result="DRAW"

    if (ch==1 and comp==2):
        print('rock wins =>',end="")
        result='rock'
    elif (ch==2 and comp==1):
        print('rock wins =>',end="")
        result='rock'


    if (ch==1 and comp==3):
        print('paper wins =>\n',end= "")
        result='paper'
    elif (ch==3 and comp==1):
        print('paper wins =>\n',end= "")
        result='paper'

    if (ch==2 and comp==3):
        print('Scissors wins =>',end="")
        result='scissor'
    elif (ch==3 and comp==2):
        print('Scissors wins =>',end="")
        result='scissor'
    if result == 'DRAW':
        print("*** Its a tie***")
    if result == choice:
        print("***User wins***")
        user=user+1
    else:
        print("***Computer wins***")
        computer=computer+1
    print("\n\nDo you want to play another round? (yes/no)")

    ans = input()
    if ans =='no':
        break

#Print total scores for user and computer
print("user scores ",user)
print("computer scores ",computer)

#Check the winner
if(user > computer):
    print("Hurray! User wins")
elif(user< computer):
    print("Computer wins")
else:
    print("Tie")
```
